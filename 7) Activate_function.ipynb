```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

### Step function

```python
x = np.linspace(-4,4, 100 )

def step_fun(x):
    x[x>=0] = 1
    x[x<0] = 0
    return x
plt.ylim((-4,4))
plt.grid()
plt.plot(x,step_fun(x.copy()))
```

Logistic

```python
def logistic(x):
    return 1/(1+np.exp(-x))
y = logistic(x.copy())
plt.ylim((-4,4))
plt.grid()
plt.plot(x,y)
```

### Tanh

```python
def tanh(x):
    return (2/(1+np.exp(-2*x))) -1

y = tanh(x.copy())
plt.ylim((-4,4))
plt.grid()
plt.plot(x,y)
```

### Relu

```python
def relu(x):
    x[x<=0] = 0
    return x

y = relu(x.copy())
plt.ylim((-4,4))
plt.grid()
plt.plot(x,y)

```

### Leaky Relu
try to solve the vanishing graient problem

```python
x = np.linspace(-4,4, 100 )
def l_relu(x, a=0.01):
    x[x<=0] = a * x[x<=0]
    return x
```

```python
y = l_relu(x.copy())
plt.ylim((-4,4))
plt.plot(x,y)
```
